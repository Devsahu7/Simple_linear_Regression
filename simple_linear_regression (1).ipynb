```python
from sklearn.linear_model import LinearRegression
# import the LinearRegression from sklearn
```

```python
model = LinearRegression()
#  here we make a only model
#  it means not train data 
```

```python
model
```

```python
model.fit(x_train,y_train)   
#  in this we train the model 
# in this part we find m and c value 
```

```python
pridict_value = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```
